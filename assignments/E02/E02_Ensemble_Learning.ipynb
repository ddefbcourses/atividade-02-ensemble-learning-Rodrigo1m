# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `Fashion MNIST`
Realize a separação em treino e teste
Utilize `train_test_split` com controle de aleatoriedade
Retorne: `X_train`, `X_test`, `y_train`, `y_test`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

````python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

def load_data(seed):
    X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)
    y = y.astype(int)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)
    return X_train, X_test, y_train, y_test
````

````python
X_train, X_test, y_train, y_test = load_data(seed=42)
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
````

Nao precisa normalizar pq arvores nao ligam pra escala. Mas e bom pra outros modelos.

# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`
`train_adaboost(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

````python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

def train_random_forest(X_train, y_train, seed):
    model = RandomForestClassifier(random_state=seed)
    model.fit(X_train, y_train)
    return model

def train_adaboost(X_train, y_train, seed):
    model = AdaBoostClassifier(random_state=seed)
    model.fit(X_train, y_train)
    return model
````

````python
X_train_small = X_train[:100]
y_train_small = y_train[:100]
model_rf = train_random_forest(X_train_small, y_train_small, seed=42)
model_ab = train_adaboost(X_train_small, y_train_small, seed=42)
print("Modelo Random Forest treinado")
print("Modelo AdaBoost treinado")
````

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf` ou `ab`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
def run_pipeline(model_type="rf", seed=42):
    X_train, X_test, y_train, y_test = load_data(seed)
    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed)
    elif model_type == "ab":
        model = train_adaboost(X_train, y_train, seed)
    else:
        raise ValueError("model_type must be 'rf' or 'ab'")
    accuracy = evaluate(model, X_test, y_test)
    return accuracy
````

````python
accuracy_rf = run_pipeline(model_type="rf", seed=42)
print(f"Acurácia Random Forest: {accuracy_rf}")
````

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

Overfitting comeca em profundidades >10. Com max_depth=None, arvore fica 100% no treino pq memoriza tudo, mas overfit.
